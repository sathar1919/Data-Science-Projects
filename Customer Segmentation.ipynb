Customer segmentation divides customers into groups based on shared
characteristics, enabling targeted marketing strategies.
Key Steps:
1. Data Collection: Gather customer data (e.g., demographics,
purchase history, behavior).
2. Data Preprocessing: Normalize data and handle missing
values.
3. Clustering: Use algorithms like K-Means or DBSCAN to
group customers.
4. Visualization: Visualize clusters to interpret results.

```python
#Segmenting Retail Customers
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Age': [25, 45, 35, 50, 23, 40, 60, 48, 33, 55], 
        'AnnualIncome (k$)': [30, 70, 50, 90, 20, 80, 100, 85, 40, 95],
        'SpendingScore (1-100)': [50, 60, 70, 80, 30, 90, 20, 85, 40, 75]
        }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)
```

```python
#Visualize clusters
plt.scatter(df['AnnualIncome (k$)'], df['SpendingScore (1-100)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation')
plt.show()
```
